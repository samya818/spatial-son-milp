## ⚖️ Le concept de Boucle Fermée (Closed-Loop)
Contrairement à une évaluation statique classique (Backtest), la boucle fermée simule l'interaction réelle entre le système SON et le réseau.

### 🔄 Le Cycle de Feedback
1. **T = 0** : Le modèle ML prédit la congestion à **T + 60min**.
2. **T = 1** : Le MILP décide d'une nouvelle configuration (Offset A3).
3. **T = 2** : Le trafic réel est observé. **Mais ce trafic est modifié** par la décision prise à T=1.
4. **T = 3** : Les nouveaux lags injectés dans le modèle ML pour la prédiction suivante tiennent compte de cette modification.

## 📊 Politiques comparées
- **Politique STATIQUE** : Pas de SON. On subit la congestion telle qu'elle arrive.
- **Politique DYNAMIQUE** : Notre système SON agit à chaque slot pour délester les zones saturées.

## 📈 Résultats observés
La politique dynamique permet de réduire le volume non satisfait de plus de **35%**, validant ainsi la robustesse du couplage ML + MILP.
